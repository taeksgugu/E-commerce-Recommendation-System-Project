#### 1. 2030 남성 -> 2030 (이유는 남성까지 넣어버리면 연관성 분석 x)

```python
Smalllst_2030male = []
df_2030male = df[(df.Gender == '남성')&(df.AgeGroup.isin(['20대', '30대']))]
for orderid in df_2030male.idOrder.unique().tolist():
    dic = {}
    dic['idOrder'] = orderid
    dic['product'] = df_2030male[(df_2030male['idOrder'] == orderid)]['ItemSmallName'].values.tolist()
    Smalllst_2030male.append(dic)
Smalldf_2030male = pd.DataFrame(Smalllst_2030male)
Smalldf_2030male
```

```python
list_association_2030male = []
for i in range(len(Smalldf_2030male)):
  list_association_2030male.append(Smalldf_2030male.loc[i,'product'])
```

```python
enc_2030male = TransactionEncoder()
df_raw_enc_2030male = enc_2030male.fit_transform(X = list_association_2030male)
df_asso_2030male = pd.DataFrame(df_raw_enc_2030male, columns = enc_2030male.columns_)
df_asso_2030male.head()
```

```python
v_min_support_2030male = 0.05
df_freq_2030male = apriori(df_asso_2030male, min_support = v_min_support_2030male, use_colnames =True)
df_asso_rule_2030male = association_rules(df_freq_2030male, metric='confidence', min_threshold = 0)
df_asso_rule_2030male.sort_values(by='confidence', ascending=False, inplace=True)
df_asso_rule_2030male.round(3)
### 없음.... 
### 대체로 2030 남성이 구매할 경우 2030 전체 데이터 기반으로 연관 분석 결과를 기반 추천할 예정정
```
